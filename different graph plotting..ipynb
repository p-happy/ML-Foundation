```python
import matplotlib.pyplot as plt 
x=["a","b","c","d"]
y=[3,5,8,7]
z=[9,8,6,5]
plt.xlabel("no of student",fontsize=20)
plt.ylabel("marks")
plt.title("students")

plt.bar(x,y,color="y",width=0.5,label="nielit")
plt.bar(x,z,color="r",width=0.5,label="nielit1")
plt.legend()
plt.show()
```
